```python
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeClassifier
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
df = pd.read_csv("documents/data/processed_data.csv")
df = df.drop(columns=["y2", "cv_fold"], axis=1)
scaler = MinMaxScaler()
df[["x1", "x2", "x3", "x4", "x5"]] = scaler.fit_transform(df[["x1", "x2", "x3", "x4", "x5"]])
X = df.drop(columns=["y1"], axis=1).to_numpy()
y = df["y1"].to_numpy()
n_alphas = 200
alphas = np.logspace(-4, 4, n_alphas)

coefs = []
for a in alphas:
    ridge = RidgeClassifier(alpha=a, fit_intercept=False, class_weight="balanced")
    ridge.fit(X, y)
    coefs.append(ridge.coef_)


ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale("log")
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel("alpha")
plt.ylabel("weights")
plt.title("Ridge Coefficients vs Regularization Strength (alpha)")
plt.axis("tight")
plt.legend(
    [f"Feature {i + 1}" for i in range(X.shape[1])], loc="best", fontsize="small"
)
plt.show()
```

```python
import plotly.graph_objects as go
import nbformat
print(nbformat.__version__)
fig = go.Figure()
coefs = np.array(coefs)  # shape (n_alphas, n_features)


for i in range(X.shape[1]):
    fig.add_trace(go.Scatter(
        x=alphas,
        y=coefs[:,i],               # each row corresponds to one feature
        mode='lines',
        name=f'Feature {i+1}',       # appears in legend and hover
        hovertemplate='Alpha: %{x}<br>Coef: %{y}<br>Feature: Feature ' + str(i+1)
    ))

# Reverse the x-axis (like your Matplotlib version)
fig.update_xaxes(type="log", autorange="reversed")

fig.update_layout(
    xaxis_title="alpha",
    yaxis_title="weights",
    title="Ridge Coefficients vs Regularization Strength (alpha)",
    template="plotly_white",
    legend_title_text="Features",
)

fig.show(renderer="browser")
```
